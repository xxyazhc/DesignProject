```python
import pandas as pd
import numpy as np
file_names = ["Eolien_Type4.txt", "Eolien_Type10.txt", "Eolien_Type21.txt"]
turbine_data = {}

for file in file_names:
    data = pd.read_csv(file, sep='\t', header=0, 
                       names=["Vitess_Vent_[m/s]", "Dir_Vent_[deg]", "Pelec[w]", 
                              "Patm[Pa]", "Temp[degC]", "Statut"])
    
    turbine_key = file.split('.')[0]
    turbine_data[turbine_key] = data
    
for turbine, data in turbine_data.items():
    print(f"\nData for {turbine}:")
    print(data.head())
```

```python
diameter_turbine4 = 150
radius_turbine4 = diameter_turbine4 / 2
area_turbine4 = 3.14159 * radius_turbine4 ** 2

turbine4_data = turbine_data["Eolien_Type4"]

filtered_turbine4_data = turbine4_data[turbine4_data["Pelec[w]"] != 0]

mean_wind_speed4 = filtered_turbine4_data["Vitess_Vent_[m/s]"].mean()
mean_pressure4 = filtered_turbine4_data["Patm[Pa]"].mean()
mean_temperature4 = filtered_turbine4_data["Temp[degC]"].mean()
mean_temperature4K = mean_temperature4+273 
mean_power4 = filtered_turbine4_data["Pelec[w]"].mean()
mean_density4 = mean_pressure4/(287*mean_temperature4K)
theoretical_power4 = (16/27)*(mean_density4)*(1/2)*(mean_wind_speed4**3)*(area_turbine4)
efficiency4 = 100*(mean_power4/theoretical_power4)

print("Mean data for Turbine 4")
print(f"Mean Wind Speed for Turbine 4: {mean_wind_speed4:.4f} m/s")
print(f"Mean Pressure for Turbine 4: {mean_pressure4:.4f} Pa")
print(f"Mean Temperature for Turbine 4: {mean_temperature4:.4f} °C")
print(f"Mean temperature for Turbine 4: {mean_temperature4K:.4f} K")
print(f"Mean Power for Turbine 4: {mean_power4:.4f} W")
print(f"Mean density for Turbine 4: {mean_density4:.4f} kg/m^3")
print(f"Theoretical power for Turbine 4: {theoretical_power4:.4f} W")
print(f"Efficiency for Turbine 4: {efficiency4:.4f} %")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

turbine_data = {
    "Eolien_Type4": turbine_data["Eolien_Type4"],
    "Eolien_Type10": turbine_data["Eolien_Type10"],
    "Eolien_Type21": turbine_data["Eolien_Type21"]
}

diameters = {
    "Eolien_Type4": 150,
    "Eolien_Type10": 110,
    "Eolien_Type21": 50
}

efficiencies = {}
mean_wind_speeds = {}
mean_powers = {}
mean_pressures = {}

for turbine, diameter in diameters.items():
    radius = diameter / 2
    area = 3.14159 * radius ** 2

    data = turbine_data[turbine]

    filtered_data = data[data["Pelec[w]"] != 0]

    mean_wind_speed = filtered_data["Vitess_Vent_[m/s]"].mean()
    mean_pressure = filtered_data["Patm[Pa]"].mean()
    mean_temperature = filtered_data["Temp[degC]"].mean()
    mean_temperatureK = mean_temperature + 273
    mean_power = filtered_data["Pelec[w]"].mean()
    mean_density = mean_pressure / (287 * mean_temperatureK)

    theoretical_power = (16/27) * (mean_density) * (1/2) * (mean_wind_speed ** 3) * area

    efficiency = 100 * (mean_power / theoretical_power)

    efficiencies[turbine] = efficiency
    mean_wind_speeds[turbine] = mean_wind_speed
    mean_powers[turbine] = mean_power
    mean_pressures[turbine] = mean_pressure

    print(f"\nMean data for {turbine}:")
    print(f"Mean Wind Speed: {mean_wind_speed:.4f} m/s")
    print(f"Mean Pressure: {mean_pressure:.4f} Pa")
    print(f"Mean Temperature: {mean_temperature:.4f} °C")
    print(f"Mean Temperature (K): {mean_temperatureK:.4f} K")
    print(f"Mean Power: {mean_power:.4f} W")
    print(f"Mean Density: {mean_density:.4f} kg/m^3")
    print(f"Theoretical Power: {theoretical_power:.4f} W")
    print(f"Efficiency: {efficiency:.4f} %")

plt.figure(figsize=(8, 6))
bars = plt.bar(efficiencies.keys(), efficiencies.values(), color=['blue', 'green', 'orange'])
plt.title("Efficiency Comparison of Wind Turbines")
plt.xlabel("Turbine Type")
plt.ylabel("Efficiency (%)")
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,                             
        f'{height:.2f}%',                   
        ha='center',                        
        va='bottom'                        
    )


plt.show()
```
